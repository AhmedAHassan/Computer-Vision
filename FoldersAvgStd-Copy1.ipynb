Joao df for 100 files with four down scales and average for each row at each size

```python
os.chdir(r"C:\Users\hp\Videos\Step1-fixedDistance_avg,std\Folders\2016-11-22-14-09-37-Joao_onlythermalthermal_image-txt")
```

```python
joao_files = []  #Joao 

for idx,file in enumerate(os.listdir()):
    with open(file,encoding="utf8", errors='ignore')as f:
        text_file = np.loadtxt(f)
        text_file[text_file<32.0]=0
        text_file[text_file>37.0]=0
        cam_res1 = cv.resize(text_file, (191,144), interpolation = cv.INTER_NEAREST)
        
        #print(np.max(cam_res1), np.min(cam_res1[cam_res1 != 0 ]))
        
        
        cam_res2 = cv.resize(cam_res1, (95,72), interpolation = cv.INTER_NEAREST)
        cam_res3 = cv.resize(cam_res2, (32,24), interpolation = cv.INTER_NEAREST)
        #rint(np.max(cam_res3), np.min(cam_res3[cam_res3 != 0 ]))
        """"
        plt.imshow(text_file,cmap='jet')
        plt.show()
        plt.imshow(cam_res1,cmap='jet')
        plt.show()
        plt.imshow(cam_res2,cmap='jet')
        plt.show()
        plt.imshow(cam_res3,cmap='jet')
        plt.show()
        print()
        """
        
        joao_files.append({"file_name":file,
                              "AvgTemp(382,288)": np.average(text_file[text_file != 0]),
                              "AvgTemp(191,144)": np.average(cam_res1[cam_res1 != 0]),
                              "AvgTemp(95,72)": np.average(cam_res2[cam_res2 != 0]),
                              "AvgTemp(32,24)": np.average(cam_res3[cam_res3 != 0])})
        
    
    if idx == 99:
        break
```

```python
df_joao = pd.DataFrame(joao_files, columns = ['file_name', "AvgTemp(382,288)","AvgTemp(191,144)","AvgTemp(95,72)","AvgTemp(32,24)"])
```
